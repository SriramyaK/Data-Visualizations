```python
import requests
```

```python
def raw_random_wiki_page():
    """Pick a random wiki page from web"""
    raw_random_wiki_page = requests.get('https://en.wikipedia.org/wiki/Special:Random')
    return raw_random_wiki_page
```

```python
def wiki_title(raw_random_wikipedia):
    """Extract title from a random Wiki page"""
    page = raw_random_wikipedia.text
    start_node = page.find("title>")
    end_node = page.find("</title>")
    wiki_title1 = page[start_node + 6: end_node]
    return wiki_title1
```

```python
wiki_title_list = []

for i in range(10):
    wiki_title_list.append(wiki_title(raw_random_wiki_page()))

print(wiki_title_list)
```
